```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### Parameters and Utility function

```python
# 3-5-5-5-3 kernel
kernel = [[-2, -1], [-2, 0], [-2, 1],
[-1, -2], [-1, -1], [-1, 0], [-1, 1], [-1, 2],
[ 0, -2], [ 0, -1], [ 0, 0], [ 0, 1], [ 0, 2],
[ 1, -2], [ 1, -1], [ 1, 0], [ 1, 1], [ 1, 2],
          [ 2, -1], [ 2, 0], [ 2, 1]]
```

```python
def valid(indices, shape):
    return (indices[0] >= 0 and indices[0] < shape[0] and 
            indices[1] >= 0 and indices[1] < shape[1])
```

### Original image

```python
img = cv2.imread('lena.bmp', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.show()
```

### Binary image

```python
def binarize(img):
    return (img >= 128) * 255
```

```python
binary_img = binarize(img)
plt.imshow(binary_img, cmap='gray')
plt.show()
```

### Dilation

```python
def template_func(in_img, out_img, kernel, func):
    n_rows, n_cols = in_img.shape
    condition = 255 if dilation else 0
        
    for i in range(n_rows):
        for j in range(n_cols):
            for k in kernel:
                u, v = k
                func(in_img, out_img, i, j, u, v)
    
    return out_img
```

```python
def dilation(img, kernel):
    dilated_img = np.zeros(img.shape, dtype=int)
    def dilation_func(in_img, out_img, i, j, u, v):
        if valid((i+u, j+v), in_img.shape) and in_img[i+u, j+v] == 255:
            out_img[i, j] = 255
    return template_func(img, dilated_img, kernel, dilation_func)
```

```python
dilated_img = dilation(binary_img, kernel)
plt.imshow(dilated_img, cmap='gray')
plt.show()
```

### Erosion

```python
def erosion(img, kernel):
    eroded_img = np.ones(img.shape, dtype=int) * 255
    def erosion_func(in_img, out_img, i, j, u, v):
        if not valid((i+u, j+v), in_img.shape) or in_img[i+u, j+v] == 0:
            out_img[i, j] = 0
    return template_func(img, eroded_img, kernel, erosion_func)
```

```python
eroded_img = erosion(binary_img, kernel)
plt.imshow(eroded_img, cmap='gray')
plt.show()
```

### Opening

```python
def opening(img, kernel):
    return dilation(erosion(img, kernel), kernel)
```

```python
opening_img = opening(binary_img, kernel)
plt.imshow(opening_img, cmap='gray')
plt.show()
```

### Closing

```python
def closing(img, kernel):
    return erosion(dilation(img, kernel), kernel)
```

```python
closing_img = closing(binary_img, kernel)
plt.imshow(closing_img, cmap='gray')
plt.show()
```

### Hit-and-miss transform

```python
def hit_and_miss_transform(img, kernel_j, kernel_k):
    img_complement = -img + 255
    return np.logical_and(erosion(img, kernel_j) == 255, erosion(img_complement, kernel_k) == 255) * 255
```

```python
kernel_j = [[0, -1], [0, 0], [1, 0]]
kernel_k = [[-1, 0], [-1, 1], [0, 1]]
hit_and_miss_transform_img = hit_and_miss_transform(binary_img, kernel_j, kernel_k)
plt.imshow(hit_and_miss_transform_img, cmap='gray')
plt.show()
```
