```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Part 1

### Original image

```python
img = cv2.imread('lena.bmp')
plt.imshow(img)
plt.show()
```

### Upside down image

```python
def upside_down(img):
    n_rows, _, _ = img.shape
    upside_down_img = np.zeros(img.shape, dtype='uint8')
    for i in range(n_rows//2):
        upside_down_img[i] = img[n_rows-i-1]
        upside_down_img[n_rows-i-1] = img[i]
    
    return upside_down_img
```

```python
upside_down_img = upside_down(img)
plt.imshow(upside_down_img)
plt.show()
```

### Right side left image

```python
def right_side_left(img):
    _, n_cols, _ = img.shape
    right_side_left_img = np.zeros(img.shape, dtype='uint8')
    for j in range(n_cols//2):
        right_side_left_img[:, j] = img[:, n_cols-j-1]
        right_side_left_img[:, n_cols-j-1] = img[:, j]
    
    return right_side_left_img
```

```python
right_side_left_img = right_side_left(img)
plt.imshow(right_side_left_img)
plt.show()
```

### Diagonally flip image

```python
def diagonally_flip(img):
    n_rows, n_cols, _  = img.shape
    diagonally_flip_img = np.zeros(img.shape, dtype='uint8')
    for i in range(n_rows):
        for j in range(n_cols):
            diagonally_flip_img[i][j] = img[j][i]
    
    return diagonally_flip_img
```

```python
diagonally_flip_img = diagonally_flip(img)
plt.imshow(diagonally_flip_img)
plt.show()
```

## Part 2

### Rotate 45 degrees clockwise

```python
def rotate_45_degrees(img):
    n_rows, n_cols, _ = img.shape
    center = (n_rows//2, n_cols//2)
    M = cv2.getRotationMatrix2D(center, -45, 1.0)
    rotate_img = cv2.warpAffine(img, M, (n_rows, n_cols))

    return rotate_img
```

```python
rotate_img = rotate_45_degrees(img)
plt.imshow(rotate_img)
plt.show()
```

### Shrink in half

```python
def shrink_in_half(img):
    n_rows, n_cols, _ = img.shape
    shrink_in_half_img = cv2.resize(img, (n_rows//2, n_cols//2))
    
    return shrink_in_half_img
```

```python
shrink_in_half_img = shrink_in_half(img)
plt.imshow(shrink_in_half_img)
plt.show()
```

### Binarize at 128

```python
def binarize_at_128(img):
    _, binarize_img = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY)

    return binarize_img
```

```python
binarize_img = binarize_at_128(img)
plt.imshow(binarize_img)
plt.show()
```
